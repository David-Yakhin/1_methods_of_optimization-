# Метод золотого сечения

```python
import numpy as np
import matplotlib.pyplot as plt

class GoldenSectionMethod:
    def __init__(self, a, b, eps=0.01):
        self.a = a
        self.b = b
        self.eps = eps
        self.phi = (np.sqrt(5) - 1) / 2  # золотое сечение ≈ 0.618
    
    def minimize(self, f):
        """
        Метод золотого сечения для минимизации функции на отрезке [a, b]
        """
        a, b = self.a, self.b
        eps = self.eps
        
        # Начальные точки
        x1 = b - self.phi * (b - a)
        x2 = a + self.phi * (b - a)
        
        f1, f2 = f(x1), f(x2)
        
        k = 0
        history = []
        
        while abs(b - a) > eps:
            k += 1
            
            if f1 < f2:
                b = x2
                x2 = x1
                f2 = f1
                x1 = b - self.phi * (b - a)
                f1 = f(x1)
            else:
                a = x1
                x1 = x2
                f1 = f2
                x2 = a + self.phi * (b - a)
                f2 = f(x2)
            
            history.append((a, b, x1, x2, f1, f2))
            print(f'Итерация {k}: a={a:.6f}, b={b:.6f}, x1={x1:.6f}, x2={x2:.6f}')
        
        x_min = (a + b) / 2
        f_min = f(x_min)
        
        print(f'\nРезультат метода золотого сечения:')
        print(f'x_min = {x_min:.6f}')
        print(f'f(x_min) = {f_min:.6f}')
        print(f'Количество итераций: {k}')
        print(f'Точность: {abs(b - a):.8f}')
        
        self._plot_process(f, history, x_min, f_min)
        return x_min, f_min
    
    def _plot_process(self, f, history, x_min, f_min):
        """Визуализация процесса метода золотого сечения"""
        plt.figure(figsize=(12, 8))
        
        # Основной график функции
        x_plot = np.linspace(self.a, self.b, 1000)
        y_plot = [f(x) for x in x_plot]
        plt.plot(x_plot, y_plot, 'b-', linewidth=2, label='f(x)')
        
        # историю итераций
        colors = plt.cm.viridis(np.linspace(0, 1, len(history)))
        
        for i, (a, b, x1, x2, f1, f2) in enumerate(history):
            if i >= len(history) - 3:  # Только последние 3 итерации
                plt.axvspan(a, b, alpha=0.2, color=colors[i], 
                           label=f'Итер. {i+1}' if i >= len(history) - 3 else "")
                plt.plot([x1, x2], [f1, f2], 'o', color=colors[i], markersize=6)
        
        #  Результат
        plt.plot(x_min, f_min, 'ro', markersize=10, label='Найденный минимум')
        plt.axvline(x_min, color='r', linestyle='--', alpha=0.7)
        
        plt.xlabel('x')
        plt.ylabel('f(x)')
        plt.title('Метод золотого сечения')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.show()

if __name__ == "__main__":
    # Тестовая функция
    f = lambda x: (x**2 - 5*x + 3)
    
    print("="*60)
    print("МЕТОД ЗОЛОТОГО СЕЧЕНИЯ")
    print("="*60)
    
    # Параметры поиска
    a, b, eps = 0, 6, 0.001
    
    # Создаем экземпляр метода
    golden_method = GoldenSectionMethod(a, b, eps)
    
    # Минимизация
    x_min, f_min = golden_method.minimize(f)
    
    # Сравнение с точным решением
    exact_x = 2.5  # Для f(x) = x² - 5x + 3 минимум в x = 2.5
    exact_f = f(exact_x)
    
    print(f"\nСравнение с точным решением:")
    print(f"Точное: x = {exact_x:.6f}, f(x) = {exact_f:.6f}")
    print(f"Найдено: x = {x_min:.6f}, f(x) = {f_min:.6f}")
    print(f"Погрешность: {abs(x_min - exact_x):.8f}")
    
```

# Градиентный спуск

```python
class Methods_min:
    def __init__(self, a, b, eps):
        self.a = a
        self.b = b
        self.eps = eps
        # Определяем функции по умолчанию
        self.f = lambda x: (x**2 - 5*x + 3)
        self.g = lambda x: (2*x - 5)
        self.h = lambda x: 2
    
    def gradient(self, x0, learning_rate, g=None):
        """Градиентный спуск"""
        if g is None:
            g = self.g  # используем производную по умолчанию
            
        c = True
        while c:    
            x_ = x0 - learning_rate * g(x0)
            if abs(g(x_)) < self.eps:
                print(f'x0 = {x_, x0}')
                break
            else:
                x0 = x_
        return x_


methods_min = Methods_min(a, b, eps=0.01)
methods_min.gradient(x0, learning_rate=0.02)  


g_custom = lambda x: 3*x**2 - 4  
methods_min.gradient(x0, learning_rate=0.02, g=g_custom)
```

# Метод покрытий

```python
import numpy as np
import matplotlib.pyplot as plt

class CoverMethod:
    def __init__(self, a, b, eps=0.01):
        self.a = a
        self.b = b
        self.eps = eps
    
    def uniform_search(self, f, n_points=100):
        """
        Метод покрытий (равномерный поиск) на отрезке [a, b]
        """
        a, b = self.a, self.b
        
        # Создаем равномерную сетку
        x_points = np.linspace(a, b, n_points)
        f_values = [f(x) for x in x_points]
        
        # Находим точку с мин. значением фун
        min_index = np.argmin(f_values)
        x_min = x_points[min_index]
        f_min = f_values[min_index]
        
        print(f'Метод равномерного покрытия:')
        print(f'x_min = {x_min:.6f}')
        print(f'f(x_min) = {f_min:.6f}')
        print(f'Использовано точек: {n_points}')
        print(f'Шаг сетки: {(b-a)/(n_points-1):.6f}')
        
        self._plot_uniform_search(f, x_points, f_values, x_min, f_min, n_points)
        
        return x_min, f_min
    
    def adaptive_search(self, f, initial_points=20, max_iterations=15):
        """
        Адаптивный метод покрытий с последовательным сжатием области поиска
        """
        a, b = self.a, self.b
        eps = self.eps
        
        history = []
        x_current = (a + b) / 2
        
        for iteration in range(max_iterations):
            
            n_points = max(10, int(initial_points * (b - a) / (self.b - self.a)))
            
            # Равно
            x_points = np.linspace(a, b, n_points)
            f_values = [f(x) for x in x_points]
            min_index = np.argmin(f_values)
            x_min = x_points[min_index]
            
            history.append({
                'iteration': iteration + 1,
                'a': a, 'b': b, 
                'x_min': x_min, 'f_min': f(x_min),
                'n_points': n_points,
                'interval_size': b - a
            })
            
            print(f'Итерация {iteration+1}: x_min = {x_min:.6f}, '
                  f'интервал = [{a:.6f}, {b:.6f}], '
                  f'точек = {n_points}')
            
            
            shrink_factor = 0.4  # Коэффициент сжатия
            interval_size = (b - a) * shrink_factor
            a_new = max(self.a, x_min - interval_size/2)
            b_new = min(self.b, x_min + interval_size/2)
            
            a, b = a_new, b_new
            x_current = x_min
            
            if b - a < eps:
                print(f"Достигнута требуемая точность на итерации {iteration+1}")
                break
        
        self._plot_adaptive_search(f, history)
        
        return x_current, f(x_current)
    
    def sequential_refinement(self, f, stages=3):
        """
        Метод последовательного сгущения сетки
        """
        a, b = self.a, self.b
        
        points_per_stage = [10, 50, 200, 1000]  # Количество точек на каждом этапе
        
        history = []
        x_best = (a + b) / 2
        f_best = f(x_best)
        
        for stage in range(min(stages, len(points_per_stage))):
            n_points = points_per_stage[stage]
            
            # Поиск на текущей сетке
            x_points = np.linspace(a, b, n_points)
            f_values = [f(x) for x in x_points]
            min_index = np.argmin(f_values)
            x_min = x_points[min_index]
            f_min = f_values[min_index]
            
            # Обн лучший результат
            if f_min < f_best:
                x_best, f_best = x_min, f_min
            
            # Сужаем область поиска для следующей шагов
            shrink = 0.5 / (stage + 1)  # Меньше сужаем на поздних стад
            interval = (b - a) * shrink
            a = max(self.a, x_min - interval)
            b = min(self.b, x_min + interval)
            
            history.append({
                'stage': stage + 1,
                'n_points': n_points,
                'x_min': x_min,
                'f_min': f_min,
                'interval': [a, b]
            })
            
            print(f'Стадия {stage+1}: {n_points} точек, '
                  f'x_min = {x_min:.6f}, интервал = [{a:.6f}, {b:.6f}]')
        
        self._plot_sequential_refinement(f, history, x_best, f_best)
        
        return x_best, f_best
    
    def _plot_uniform_search(self, f, x_points, f_values, x_min, f_min, n_points):
        """Визуализация равномерного поиска"""
        plt.figure(figsize=(12, 6))
        
        # Основной график функции
        x_plot = np.linspace(self.a, self.b, 1000)
        y_plot = [f(x) for x in x_plot]
        plt.plot(x_plot, y_plot, 'b-', linewidth=2, label='f(x)', alpha=0.7)
        
        # Точки поиска
        plt.plot(x_points, f_values, 'ro', markersize=4, label=f'Точки поиска ({n_points} шт.)')
        plt.plot(x_min, f_min, 'go', markersize=10, label='Найденный минимум')
        plt.axvline(x_min, color='g', linestyle='--', alpha=0.7)
        
        plt.xlabel('x')
        plt.ylabel('f(x)')
        plt.title(f'Метод равномерного покрытия ({n_points} точек)')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.show()
    
    def _plot_adaptive_search(self, f, history):
        """Визуализация адаптивного поиска"""
        plt.figure(figsize=(12, 8))
        
        # Основной график функции
        x_plot = np.linspace(self.a, self.b, 1000)
        y_plot = [f(x) for x in x_plot]
        plt.plot(x_plot, y_plot, 'b-', linewidth=2, label='f(x)')
        
        # Отображаем историю поиска
        colors = plt.cm.plasma(np.linspace(0, 1, len(history)))
        
        for i, record in enumerate(history):
            a, b = record['a'], record['b']
            x_min, f_min = record['x_min'], record['f_min']
            
            plt.axvspan(a, b, alpha=0.2, color=colors[i])
            plt.plot(x_min, f_min, 'o', color=colors[i], markersize=6,
                    label=f'Итер. {i+1}' if i < 3 else "")
        
        # Результат
        final = history[-1]
        plt.plot(final['x_min'], final['f_min'], 'ro', markersize=12, 
                label='Финальный минимум')
        plt.axvline(final['x_min'], color='r', linestyle='--', alpha=0.7)
        
        plt.xlabel('x')
        plt.ylabel('f(x)')
        plt.title('Адаптивный метод покрытий')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.show()
    
    def _plot_sequential_refinement(self, f, history, x_best, f_best):
        """Визуализация последовательного сгущения"""
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
        
        # График 1: Процесс поиска
        x_plot = np.linspace(self.a, self.b, 1000)
        y_plot = [f(x) for x in x_plot]
        ax1.plot(x_plot, y_plot, 'b-', linewidth=2, label='f(x)')
        
        colors = plt.cm.viridis(np.linspace(0, 1, len(history)))
        for i, record in enumerate(history):
            a, b = record['interval']
            x_min = record['x_min']
            
            ax1.axvspan(a, b, alpha=0.2, color=colors[i])
            ax1.plot(x_min, record['f_min'], 'o', color=colors[i], markersize=6,
                    label=f'Стадия {i+1}')
        
        ax1.plot(x_best, f_best, 'ro', markersize=10, label='Лучший результат')
        ax1.set_xlabel('x')
        ax1.set_ylabel('f(x)')
        ax1.set_title('Последовательное сгущение сетки')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # График 2: Сходимость
        stages = [r['stage'] for r in history]
        errors = [abs(r['x_min'] - x_best) for r in history]
        points = [r['n_points'] for r in history]
        
        ax2.plot(stages, errors, 'bo-', linewidth=2, markersize=8, label='Погрешность')
        ax2.set_xlabel('Стадия')
        ax2.set_ylabel('Погрешность', color='b')
        ax2.tick_params(axis='y', labelcolor='b')
        ax2.grid(True, alpha=0.3)
        
        ax2_twin = ax2.twinx()
        ax2_twin.plot(stages, points, 'ro-', linewidth=2, markersize=8, label='Точек на стадии')
        ax2_twin.set_ylabel('Количество точек', color='r')
        ax2_twin.tick_params(axis='y', labelcolor='r')
        
        ax2.set_title('Сходимость метода')
        fig.legend(loc='upper center', bbox_to_anchor=(0.5, 0.05), ncol=2)
        plt.tight_layout()
        plt.show()


if __name__ == "__main__":
    # Тест функция
    f = lambda x: (x**2 - 5*x + 3)
    
    print("="*60)
    print("МЕТОДЫ ПОКРЫТИЙ")
    print("="*60)
    
    a, b, eps = 0, 6, 0.001
    cover_method = CoverMethod(a, b, eps)
    
    # 1. Равн-й поиск
    print("\n" + "="*40)
    print("1. РАВНОМЕРНЫЙ ПОИСК")
    print("="*40)
    x_uniform, f_uniform = cover_method.uniform_search(f, n_points=50)
    
    # 2. Адапт поиск
    print("\n" + "="*40)
    print("2. АДАПТИВНЫЙ ПОИСК")
    print("="*40)
    x_adaptive, f_adaptive = cover_method.adaptive_search(f, initial_points=30)
    
    # 3. Последовательное сгущение
    print("\n" + "="*40)
    print("3. ПОСЛЕДОВАТЕЛЬНОЕ СГУЩЕНИЕ")
    print("="*40)
    x_sequential, f_sequential = cover_method.sequential_refinement(f, stages=4)
    
    # Сравнение результатов
    exact_x = 2.5
    exact_f = f(exact_x)
    
    print("\n" + "="*60)
    print("СРАВНЕНИЕ РЕЗУЛЬТАТОВ")
    print("="*60)
    print(f"{'Метод':<25} {'x_min':<10} {'f(x_min)':<12} {'Погрешность':<12}")
    print("-" * 60)
    print(f"{'Точное решение':<25} {exact_x:<10.6f} {exact_f:<12.6f} {0:<12.8f}")
    print(f"{'Равномерный поиск':<25} {x_uniform:<10.6f} {f_uniform:<12.6f} {abs(x_uniform-exact_x):<12.8f}")
    print(f"{'Адаптивный поиск':<25} {x_adaptive:<10.6f} {f_adaptive:<12.6f} {abs(x_adaptive-exact_x):<12.8f}")
    print(f"{'Последовательное сгущение':<25} {x_sequential:<10.6f} {f_sequential:<12.6f} {abs(x_sequential-exact_x):<12.8f}")
```
